```python
plt.figure(figsize=(6, 4))
df['sentiment_scores'].value_counts().plot(kind='bar', color=['pink', 'skyblue'])
plt.title('Distribution of Sentiments')
plt.xlabel('sentiment_scores')
plt.ylabel('genre')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
